1) Implementar Linked List con objetos. Cada nodo de la Linked List debe almacenar un dato, este puede ser un int, un float, un string o del tipo que deseen. Se deben crear, como mínimo, los siguientes métodos los para trabajar con los datos. Pueden implementar métodos adicionales si lo necesitan o desean.

    NOTA: No pueden utilizar una lista como estructura de datos, es decir no se pueden guardar todos los datos en una lista que se utiliza como una Linked List.

- Agregar (dato) -> None: Agrega un nuevo dato a la Linked List. Se puede agregar al inicio, al final de la lista o, si se desea, agregarla de forma que la Linked List quede ordenada.
- Buscar (dato) -> Bool: Busca un dato en la Linked List. Si el dato está, devuelve True, sino, False. Alternativamente la función puede devolver el nodo que contiene el dato buscado.
- Borrar (dato) -> LinkedList: Elimina un nodo de la Linked List que contiene el dato suministrado.
- Listar () -> list[dato]: Devuelve una lista con todos los datos de la Linked List.

```python
class Nodo: # Esta clase maneja los nodos, cada instancia es un nodo
    
    def __init__(self, dato) -> None:
        self.dato = dato
        self.siguiente = None

class Linked_List: # En esta estarán los metodos para manejar la Linked_List
    
    def __init__(self) -> None: # Este es para el inicio de la Linked_list -> Sino deberia poner una variable de inicio en la ejecucion del programa
        self.inicio = None # Se comporta como el inicio de la lista (head - cabeza)
    
    def agregar_nuevo(self, nuevo):  # Método para agregar al principio
        nuevo_nodo = Nodo(nuevo)  # Creo un nodo con el dato
        nuevo_nodo.siguiente = self.inicio  # Apunto el nuevo nodo al que era el primer nodo
        self.inicio = nuevo_nodo  # Ahora el nuevo nodo es el primero de la lista

    def buscar(self, incognita): # Buscamor, hay que pasarle el valor a buscar
        nodo_analizado = self.inicio # Empezamos en analizando el último valor cargado
        
        while nodo_analizado is not None: # Recorremos los nodos hasta llegar al final (None - Cola - Tail)
            if nodo_analizado.dato == incognita: 
                return True # Cuando coincida devuelvo un True 
            nodo_analizado = nodo_analizado.siguiente # Pasamos al siguiente
            
        return False
    
    def listar(self) -> list: # Este método deberia poner en una lista de HEAD a TAIL los datos de los Nodos
        lista_datos = [] # Lista vacia para hacer append de los datos
        actual = self.inicio # Nos ubicamos en HEAD
        
        while actual is not None: # Recorremos los nodos hasta llegar al final (None - Cola - Tail)
            lista_datos.append(actual.dato) # Cada dato lo ponemos en la lista
            actual = actual.siguiente # Leemos el siguiente
        return lista_datos 

    def borrar(self, eliminar) -> int:
        nodo_analizado = self.inicio # Empezamos en analizando el último valor cargado
        auxiliar = None # Es para ir reemplazando los valores ( ya que no se borra con DEL)
        cantidad_eliminados = 0 # Cuento la cantidad de borrados
        
        while nodo_analizado is not None: # Recorremos los nodos hasta llegar al final (None - Cola - Tail)
            if nodo_analizado.dato == eliminar: # Cuando encuentro al que quiero eliminar
                if auxiliar is None: # Si estoy al principio de la lista
                    self.inicio = nodo_analizado.siguiente # guardo el siguiente en la clase
                    nodo_analizado = self.inicio # reemplazo el que estoy viendo por el siguiente
                else: # Si lo encuentro en un nodo que no es el primero
                    auxiliar.siguiente = nodo_analizado.siguiente
                    nodo_analizado = nodo_analizado.siguiente
                cantidad_eliminados += 1 # Sumo uno a la cantidad de eliminados
            else: # Si no encuentro nada que eliminar continúo recorriendo
                auxiliar = nodo_analizado
                nodo_analizado = nodo_analizado.siguiente
            
        return cantidad_eliminados


#########################################################

if __name__ == '__main__':

    # Inicio programa
    lista = Linked_List()

    # Agregar valores
    cantidad = 0

    try:
        cantidad = int(input('¿Cuántos valores quiere agregar solo números?'))
    except ValueError:
        print('Usted ingresó algo diferente a un numero entero, inicie nuevamente el programa')

    while cantidad != 0:
        try:
            lista.agregar_nuevo(int(input('Agregue un numero entero a la lista [si se equivoca colocaremos un cero o reinicie el programa]')))
        except ValueError:
            lista.agregar_nuevo(0)
        cantidad -= 1

    # Listar antes de borrar:
    print("Listado:", lista.listar())

    # Buscar:
    try:
        buscado = int(input('¿Qué número quiere buscar?'))
        resultado = lista.buscar(buscado)
        print(resultado)
    except ValueError:
        print("Usted agregó un valor erróneo")

    # Borrar
    try:
        lista.borrar(int(input('¿Qué número quiere borrar?')))
    except ValueError:
        print("Usted agrego un valor erroneo")

    # Listar despues de borrar
    print("Listado final:", lista.listar())
```

2) Implementar Binary Tree con objetos. Cada nodo del Binary Tree debe almacenar un dato, este puede ser un int, un float, un string o de cualquier tipo que implemente los métodos < y > (mayor y menor). Se deben crear, como mínimo, los siguientes métodos los para trabajar con los datos. Pueden implementar métodos adicionales si lo necesitan o desean.

- Agregar (dato) -> None: Agrega un nuevo dato al Binary Tree. Se debe agregar de forma que el Binary Tree quede ordenado.
- Buscar (dato) -> Bool: Busca un dato en la Binary Tree. Si el dato está, devuelve True, sino, False. - Alternativamente la función puede devolver el nodo que contiene el dato buscado.
- Borrar (dato) -> BinaryTree: Elimina un nodo de la Binary Tree que contiene el dato suministrado.
- Listar () -> list[dato]: Devuelve una lista con todos los datos de la Binary Tree ordenados.

```python
# Clase que guarda nodos
class Nodo:
    def __init__(self, key) -> None:
        self.left = None # aca van los menores
        self.right = None # aca van los mayores
        self.val = key # la raiz de cada nodo

# Acciones sobe nodos
class Binary:
    def __init__(self):
        self.raiz = None

    def agregar(self, nuevo): # función que crea el primer nodo
        self.dato = nuevo
        if self.raiz == None: # ¿if not self.raiz?
            self.raiz = Nodo(nuevo) # al estar vacia la raiz pone el primer dato ingresado como raíz:( None - NUEVO - None )
        else:
            self._insertar(self.raiz, nuevo) # Metodo privado - El usuario no lo usa se comienza a usar internamente una vez generado el primer nodo

    def _insertar(self, nodo_actual, nuevo): 
            if nuevo < nodo_actual.val:
                if nodo_actual.left is None:
                    nodo_actual.left = Nodo(nuevo) # ( NUEVO - Raiz - None )
                else:
                    self._insertar(nodo_actual.left, nuevo) # Si no es menor no modifico
            elif nuevo > nodo_actual.val:
                if nodo_actual.right is None:
                    nodo_actual.right = Nodo(nuevo) # ( None - Raiz - NUEVO )
                else:
                    self._insertar(nodo_actual.right, nuevo) # Si no es menor no modifico
            ###### Verificar que pasa si es igual ######

    def buscar(self, buscado):
        nodo_actual = self.raiz # inicio nodo

        while nodo_actual is not None: # si hay valor cargado
            if buscado == nodo_actual.val:
                return True # Si es igual devuelve True
            elif buscado < nodo_actual.val: 
                nodo_actual = nodo_actual.left # Si es me menor me posiciono para ahi para analizar
            else:
                nodo_actual = nodo_actual.right # Si no es menor me posiciono en el lado mayor para analizar
        return False # Cuando no encuetra devuelve False

    def listar_diccionario(self):
        diccionario_nodos = {} # dict vacio para guardar el arbol
        contador = {"nodo": 1}  # me permite no tener que hacer una variable aparte

        def recorrer(nodo):
            if nodo is None: # si no hay nada termino
                return

            numero_nodo = contador["nodo"]
            diccionario_nodos[f"Raiz{numero_nodo}"] = nodo.val # analizo el nodo x y le agrego el numero de nodo a la key raiz + el valor de la key

            if nodo.left:
                diccionario_nodos[f"Menor{numero_nodo}"] = nodo.left.val # uso el mismo numero de nodo para agregar a la key menor + el valor del nuemro menor
            if nodo.right:
                diccionario_nodos[f"Mayor{numero_nodo}"] = nodo.right.val # uso el mismo numero de nodo para agregar a la key mayor + el valor del nuemro mayor

            contador["nodo"] += 1 # sumo 1 al número del nodo

            # Recursión a izquierda y derecha
            recorrer(nodo.left) # me posiciono al principio y voy para el lado del menor
            recorrer(nodo.right) # me posiciono al principio y voy para el lado del menor

        recorrer(self.raiz) # inico del nodo
        return diccionario_nodos # Lo intente imprimir con __str__, logre avances pero los borré por que en realidad pide una lista el ejercicio

    def listar_lista(self):
        lista_nodos = [] # lista vacia para guardar los valores del arbol

        def recorrer(nodo):
            if nodo is None: # si no hay nada termino
                return
            recorrer(nodo.left) # voy para el lado del menor 
            lista_nodos.append(nodo.val) # agrego los valores sin importar del lado que este
            recorrer(nodo.right) # voy por el lado de los mayores

        recorrer(self.raiz) # inico del nodo
        return lista_nodos

    def borrar(self, borrado):
        self.raiz = self._borrar_nodo(self.raiz, borrado)  # actualiza la raíz por si se borra la misma

    def _borrar_nodo(self, nodo, borrado):
        if nodo is None:
            return None # si no hay nada termino

        if borrado < nodo.val: # 
            nodo.left = self._borrar_nodo(nodo.left, borrado)  # voy por el lado del menor
        elif borrado > nodo.val:
            nodo.right = self._borrar_nodo(nodo.right, borrado)  # voy por el lado mayor
        else:
            # cuando no tiene menor ni mayor
            if nodo.left is None and nodo.right is None:
                return None # No hago nada mas
            # cuando tiene un menor o un mayor
            elif nodo.left is None:
                return nodo.right # si hay menor devuelvo el mayor
            elif nodo.right is None:
                return nodo.left # si hay mayor devuelvo menor
            # cuando tiene un menor y un mayor
            else:
                sucesor = self._encontrar_minimo(nodo.right)  # busco el menor del lado derecho
                nodo.val = sucesor.val  # reemplazo valor
                nodo.right = self._borrar_nodo(nodo.right, sucesor.val)  # borro el sucesor

        return nodo

    def _encontrar_minimo(self, nodo): # busco el minimo 
        while nodo.left is not None:
            nodo = nodo.left
        return nodo
##########################################################################################################################################################

if __name__ == "__main__":

    # Inicio programa
    arbol = Binary()

    # Agregar valores
    cantidad = 0

    try:
        cantidad = int(input('¿Cuántos valores quiere agregar solo números?'))
    except ValueError:
        print('Usted ingresó algo diferente a un numero entero, inicie nuevamente el programa')

    while cantidad != 0:
        try:
            arbol.agregar(int(input('Agregue un numero entero a la lista ')))
        except ValueError:
            print("Usted agrego un valor no permitido, empiece de nuevo")
        cantidad -= 1

    # Imprimo antes de borrar
    print(f"Binary tree completo: {arbol.listar_lista()}")
    
    # Buscar Valor
    try:
        buscado = int(input('¿Qué número quiere buscar?'))
        resultado = arbol.buscar(buscado)
        print(resultado)
    except ValueError:
        print(f"Usted agregó un valor no permitido")

    # Borrar valor:
    try:
        borrado = int(input('¿Qué número quiere borrar?'))
        resultado = arbol.borrar(borrado)
    except ValueError:
        print(f"Usted agregó un valor no permitido")

    # print(arbol.listar_diccionario()) # DEBUG anda bien. Ver nota mas arriba
    
    print(f"Binary tree despues de borrar: {arbol.listar_lista()}")
```
