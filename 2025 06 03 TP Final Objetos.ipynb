3) Crear una clase Agenda que utilice los objetos de la clase Tarea y Evento de los ejercicios anteriores. Implementar métodos para agregar y eliminar tareas y eventos. Listar las tareas por orden de prioridad y por orden de fecha de vencimiento. Listar eventos por fecha o por ubicación. Se pueden agregar más atributos y/o métodos que crean convenientes.

```python
# Como en los ejercicios anteriores copie muchos elementos de un código al otro, voy a intentar usar herencia para los atributos y métodos comunes.
from datetime import datetime, timedelta
from tkinter import messagebox

class Elemento_agenda: # Guardo los elementos comunes / Uso el poder de la herencia para resumir mucho código.
    def __init__(self, nombre, descripcion, fecha_limite):
        self.nombre = nombre
        self.estado = None # True = Finalizado/Asistido /// False = Vencida/Ya Pasó /// None = Activo
        self.tiempo_restante = None
        self.descripcion = descripcion
        self.fecha_limite = fecha_limite
    
    def completar(self):
        self.estado = True
    
    def vencio(self, fecha_actual=None):
        if fecha_actual is None:
            fecha_actual = datetime.now()
        if self.fecha_limite < fecha_actual:
            if self.estado != True: # lo agregue por error al modificar fechas
                self.estado = False # Indica que está vencido
            self.tiempo_restante = None
        else:
            self.tiempo_restante = self.fecha_limite - fecha_actual # Mustra cuanto tiempo le quedá a la tarea
            if self.estado != True: # lo agregue por error al modificar fechas
                self.estado = None # lo agregue antes que el if pero me siguio generando problemas al completar
    
    def modificar(self, atributo_modificar, valor_nuevo):
        if atributo_modificar == 'vencimiento': # Otra correcion para evitar errores en la ejecución puesta luego de varias pruebas
            atributo_modificar = 'fecha_limite'
        if hasattr(self, atributo_modificar):
            setattr(self, atributo_modificar, valor_nuevo)
        if atributo_modificar == 'fecha_limite':
            self.vencio()

class Tarea(Elemento_agenda):
    
    def __init__(self, nombre, descripcion, vencimiento, prioridad):
        super().__init__(nombre, descripcion, vencimiento) # LLamo los atributos desde la clase Elemento_agenda
        self.prioridad = prioridad

class Evento(Elemento_agenda):
    
    def __init__(self, nombre, descripcion, fecha, duracion, ubicacion):
        super().__init__(nombre, descripcion, fecha) # LLamo los atributos desde la clase Elemento_agenda
        self.ubicacion = ubicacion
        self.duracion = duracion

class Agenda:
    def __init__(self):
        self.lista_tareas = []
        self.lista_eventos = []
    
    def agregar_tarea(self, nombre, descripcion, vencimiento, prioridad):
        
        vencimiento_texto = datetime.strptime(vencimiento, "%d/%m/%Y") # convierte fecha en texto en fecha
        
        if prioridad == "alta":
            prioridad_mod = 1
        elif prioridad == "normal":
            prioridad_mod = 2
        elif prioridad == "baja":
            prioridad_mod = 3
        else:
            prioridad_mod = 2  # Por defecto
        
        nueva_tarea = Tarea(nombre, descripcion, vencimiento_texto, prioridad_mod)
        self.lista_tareas.append(nueva_tarea)
    
    def agregar_evento(self, nombre, descripcion, fecha, duracion, ubicacion):
        
        fecha_texto = datetime.strptime(fecha, "%d/%m/%Y") # convierte fecha en texto en fecha
        
        nuevo_evento = Evento(nombre, descripcion, fecha_texto, duracion, ubicacion)
        self.lista_eventos.append(nuevo_evento)
    
    def borrar_elemento(self, elemento_borrar, lista_elemento):
        for t in lista_elemento:
            if t.nombre.lower() == elemento_borrar.lower():
                lista_elemento.remove(t)
                return t.nombre

    def buscar_elemento(self, nombre_buscado, lista_elemento): # Busco un evento en la lista
        for e in lista_elemento:
            if e.nombre.lower() == nombre_buscado.lower():
                return e
        return None
    
    def listar_tareas(self, ordenar_por): # No sé si era válido usar sorted, hace rato lo vengo leyendo y probado. Ademas se me complicó demasiado el programa como para reordenar la lista con for
        if ordenar_por.lower() == 'prioridad':
            tareas_por_prioridad = sorted(self.lista_tareas, key=lambda tarea: tarea.prioridad)
            return tareas_por_prioridad
        elif ordenar_por.lower() == 'vencimiento':
            tareas_por_vencimiento = sorted(self.lista_tareas, key=lambda tarea: tarea.fecha_limite)
            return tareas_por_vencimiento
    
    def listar_eventos(self, ordenar_por):
        if ordenar_por.lower() == 'fecha':
            eventos_por_fecha = sorted(self.lista_eventos, key=lambda evento: evento.fecha_limite)
            return eventos_por_fecha
        elif ordenar_por.lower() == 'ubicacion':
            eventos_por_ubicacion = sorted(self.lista_eventos, key=lambda evento: evento.ubicacion)
            return eventos_por_ubicacion
```

```python
# Como en los ejercicios anteriores copie muchos elementos de un código al otro, voy a intentar usar herencia para los atributos y métodos comunes.
from datetime import datetime, timedelta
from tkinter import messagebox

class Elemento_agenda: # Guardo los elementos comunes / Uso el poder de la herencia para resumir mucho código.
    def __init__(self, nombre, descripcion, fecha_limite):
        self.nombre = nombre
        self.estado = None # True = Finalizado/Asistido /// False = Vencida/Ya Pasó /// None = Activo
        self.tiempo_restante = None
        self.descripcion = descripcion
        self.fecha_limite = fecha_limite
    
    def completar(self):
        self.estado = True
    
    def vencio(self, fecha_actual=None):
        if fecha_actual is None:
            fecha_actual = datetime.now()
        if self.fecha_limite < fecha_actual:
            if self.estado != True: # lo agregue por error al modificar fechas
                self.estado = False # Indica que está vencido
            self.tiempo_restante = None
        else:
            self.tiempo_restante = self.fecha_limite - fecha_actual # Mustra cuanto tiempo le quedá a la tarea
            if self.estado != True: # lo agregue por error al modificar fechas
                self.estado = None # lo agregue antes que el if pero me siguio generando problemas al completar
    
    def modificar(self, atributo_modificar, valor_nuevo):
        if atributo_modificar == 'vencimiento': # Otra correcion para evitar errores en la ejecución puesta luego de varias pruebas
            atributo_modificar = 'fecha_limite'
        if hasattr(self, atributo_modificar):
            setattr(self, atributo_modificar, valor_nuevo)
        if atributo_modificar == 'fecha_limite':
            self.vencio()

class Tarea(Elemento_agenda):
    
    def __init__(self, nombre, descripcion, vencimiento, prioridad):
        super().__init__(nombre, descripcion, vencimiento) # LLamo los atributos desde la clase Elemento_agenda
        self.prioridad = prioridad

class Evento(Elemento_agenda):
    
    def __init__(self, nombre, descripcion, fecha, duracion, ubicacion):
        super().__init__(nombre, descripcion, fecha) # LLamo los atributos desde la clase Elemento_agenda
        self.ubicacion = ubicacion
        self.duracion = duracion

class Agenda:
    def __init__(self):
        self.lista_tareas = []
        self.lista_eventos = []
    
    def agregar_tarea(self, nombre, descripcion, vencimiento, prioridad):
        
        vencimiento_texto = datetime.strptime(vencimiento, "%d/%m/%Y") # convierte fecha en texto en fecha
        
        if prioridad == "alta":
            prioridad_mod = 1
        elif prioridad == "normal":
            prioridad_mod = 2
        elif prioridad == "baja":
            prioridad_mod = 3
        else:
            prioridad_mod = 2  # Por defecto
        
        nueva_tarea = Tarea(nombre, descripcion, vencimiento_texto, prioridad_mod)
        self.lista_tareas.append(nueva_tarea)
    
    def agregar_evento(self, nombre, descripcion, fecha, duracion, ubicacion):
        
        fecha_texto = datetime.strptime(fecha, "%d/%m/%Y") # convierte fecha en texto en fecha
        
        nuevo_evento = Evento(nombre, descripcion, fecha_texto, duracion, ubicacion)
        self.lista_eventos.append(nuevo_evento)
    
    def borrar_elemento(self, elemento_borrar, lista_elemento):
        for t in lista_elemento:
            if t.nombre.lower() == elemento_borrar.lower():
                lista_elemento.remove(t)
                return t.nombre

    def buscar_elemento(self, nombre_buscado, lista_elemento): # Busco un evento en la lista
        for e in lista_elemento:
            if e.nombre.lower() == nombre_buscado.lower():
                return e
        return None
    
    def listar_tareas(self, ordenar_por): # No sé si era válido usar sorted, hace rato lo vengo leyendo y probado. Ademas se me complicó demasiado el programa como para reordenar la lista con for
        if ordenar_por.lower() == 'prioridad':
            tareas_por_prioridad = sorted(self.lista_tareas, key=lambda tarea: tarea.prioridad)
            return tareas_por_prioridad
        elif ordenar_por.lower() == 'vencimiento':
            tareas_por_vencimiento = sorted(self.lista_tareas, key=lambda tarea: tarea.fecha_limite)
            return tareas_por_vencimiento
    
    def listar_eventos(self, ordenar_por):
        if ordenar_por.lower() == 'fecha':
            eventos_por_fecha = sorted(self.lista_eventos, key=lambda evento: evento.fecha_limite)
            return eventos_por_fecha
        elif ordenar_por.lower() == 'ubicacion':
            eventos_por_ubicacion = sorted(self.lista_eventos, key=lambda evento: evento.ubicacion)
            return eventos_por_ubicacion

#############################################################################################################################
# Funciones complementarias

# Menu "interactivo":
def mostrar_menu():
    print("\n--- MENÚ DE TAREAS ---")
    print("1 - Agregar tarea")
    print("2 - Modificar tarea")
    print("3 - Marcar tarea como completada")
    print("4 - Buscar tarea")
    print("5 - Borrar tarea")
    print("6 - Listar tareas por prioridad")
    print("7 - Listar tareas por vencimiento")
    print("\n--- MENÚ DE EVENTOS ---")
    print("A - Agregar evento")
    print("B - Modificar evento")
    print("C - Marcar evento como asistido")
    print("D - Buscar evento")
    print("E - Borrar evento")
    print("F - Listar eventos por fecha")
    print("G - Listar eventos por ubicacion")
    print("X - Salir")
    return input("Seleccione una opción: ")

# Al transformar elementos como prioridad y estado, para que sean mas genéricos o el sorted los ordene, al listarlos quedaba mal. Así que corregi eso con el get y un diccionario.

def traducir_prioridad(valor):
    return {1: "Alta", 2: "Normal", 3: "Baja"}.get(valor, "Desconocida")

def traducir_estado(valor):
    return {True: "Completada / Asistido", False: "Vencida / Pasó", None: "Activa"}.get(valor, "Desconocido")

#############################################################################################################################
## Inicio programa:

if __name__ == "__main__":
    
    agenda = Agenda()

    # Cargo cosas para que no esten las listas vacias:
    # Precarga tareas
    agenda.agregar_tarea("comprar leche", "Ir al super y comprar leche", "25/06/2023", "alta")
    agenda.agregar_tarea("estudiar python", "Practicar herencia y clases", "30/06/2025", "normal")
    agenda.agregar_tarea("pagar internet", "Vence el pago de la factura", "22/05/2025", "alta")
    agenda.agregar_tarea("lavar ropa", "Lavar ropa blanca", "21/06/2025", "baja")
    agenda.agregar_tarea("leer libro", "Leer 20 páginas del libro actual", "28/06/2025", "normal")

    # Precarga eventos
    agenda.agregar_evento("cumple de Juan", "Festejo de cumpleaños en su casa", "26/06/2025", "4", "Calle Falsa 123")
    agenda.agregar_evento("entrevista laboral", "Entrevista con empresa X", "24/06/2025", "1", "Oficina central")
    agenda.agregar_evento("clase yoga", "Clase de yoga semanal", "22/06/2025", "1.5", "Gimnasio Salud Total")
    agenda.agregar_evento("asamblea consorcio", "Reunión mensual del edificio", "27/06/2025", "2", "Zoom")
    agenda.agregar_evento("viaje al campo", "Día de paseo y descanso", "29/06/2022", "8", "Campo de los abuelos")

    messagebox.showinfo(
    message= f'{"Bienvenido/a a su Agenda. Por favor digame que desea hacer"}', 
    title="Agenda Python")
    
    # Apenas inicio el programa me fijo si vencio alguna tarea
    if agenda.lista_tareas: # Idem if agenda.lista_tareas == True, osea si tiene algo adentro
        for t in agenda.lista_tareas:
            t.vencio()
            if t.estado == False:
                print(f'Tareas vencidas\nLa tarea {t.nombre} ha vencido: {t.fecha_limite.strftime("%d/%m/%Y")}\n---------------\n') # Como es una iteracion va a dar como resultado todas las que esten vencidas
    
    # Apenas inicio el programa me fijo si vencio algun evento
    if agenda.lista_eventos:
        for e in agenda.lista_eventos:
            e.vencio()
            if e.estado == False:
                print(f'Evento vencidos\nEl evento {e.nombre} ya pasó {e.fecha_limite.strftime("%d/%m/%Y")}\n---------------\n') # Como es una iteracion va a dar como resultado todas las que esten vencidas

    while True:
        opcion = mostrar_menu() # Hice este menu para organizarme yo mejor, no se si habia una opción mejor.

        # Si el usuario elige un accion de tareas :
        if opcion == '1':
            while True:
                try: # Primero puse cada ingreso con un try pero quedaba feo el código, exception captura varios tipos de errores
                    nombre = input('¿Cómo se llama su tarea? ')
                    nombre_existe = False # me evita un bucle infinito
                    for t in agenda.lista_tareas:
                        if t.nombre.lower() == nombre.lower(): # Evita repeticiones
                            nombre_existe = True
                    if nombre_existe: # Idem nombre if nombre_existe == True
                        print('Usted ingreso un nombre que ya exite')
                    else:
                        descripcion = input('Ingrese una descripción para su tarea: ')
                        vencimiento = input('¿Cúando vence su tarea? [dd/mm/aaaa] ') 
                        prioridad = input('¿Cúal es la prioridad? [Alta / Normal / Baja]: ').lower()
                    
                        agenda.agregar_tarea(nombre, descripcion, vencimiento, prioridad)
                        print("Tarea agregada correctamente.")
                        break

                except Exception as e: # Captura errores posibles, y muestra error.
                    print("Hubo un error:", e)
                    continue
        
        elif opcion == '2': # Modificar tarea
            while True:
                try:
                    quiere_modificar = input("¿Usted quiere modificar una tarea? [ SI - NO ]")
                    if quiere_modificar.lower() != "si":
                        break
                    tarea_modificar = input('¿Cómo se llama la tarea que quiere modificar?')
                except ValueError:
                    print('Valor incorrecto')
                    continue

                tarea_encontrada = agenda.buscar_elemento(tarea_modificar, agenda.lista_tareas)
                if tarea_encontrada:
                    while True:
                        otro_atributo = input("¿Usted quiere modificar un atributo? [ SI - NO ] ")
                        if otro_atributo.lower() != "si":
                            break
                        
                        atributo = input("¿Qué atributo desea modificar? (nombre, descripcion, vencimiento, prioridad): ")
                        nuevo_valor = input("Nuevo valor: ")
                        
                        if atributo == "vencimiento":
                            try:
                                nuevo_valor = datetime.strptime(nuevo_valor, "%d/%m/%Y")
                                atributo = "fecha_limite" # Parche para problema numero 999... al modificar
                            except:
                                print("Fecha inválida. No se modificó.")
                                continue
                        elif atributo == "prioridad":
                            if nuevo_valor.lower() == "alta":
                                nuevo_valor = 1
                            elif nuevo_valor.lower() == "normal":
                                nuevo_valor = 2
                            elif nuevo_valor.lower() == "baja":
                                nuevo_valor = 3
                            else:
                                print("Prioridad inválida.")
                                continue
                        
                        tarea_encontrada.modificar(atributo, nuevo_valor) # Descripcion y  nombre no necesitan ser procesados, asi que se ingresan directamente con el metodo modificar.
                        print(f'El atributo {atributo} de la tarea {tarea_modificar} fue modificado correctamente')
        
        elif opcion == '3': # completar
            while True:
                marcar_completada = input('¿Quiere marcar una tarea como completada? [ SI - NO ]')
                if marcar_completada.lower() != "si":
                    break
                
                try:
                    tarea_completar = input('¿Cómo se llama la tarea que quiere completar?')
                except ValueError:
                    print('Valor incorrecto')
                    continue
                tarea = agenda.buscar_elemento(tarea_completar, agenda.lista_tareas) # Si la tarea exite la marco como completada.
                if tarea:
                    tarea.completar()
                    print("Tarea marcada como completada.")
                else:
                    print("Tarea no encontrada.")
        
        elif opcion == '4': # Buscar
            while True:
                try:
                    quiere_buscar = input("¿Usted quiere buscar una tarea? [ SI - NO ]")
                    if quiere_buscar.lower() != "si":
                        break
                    tarea_buscada = input('¿Cómo se llama la tarea que quiere buscar?')
                except ValueError:
                    print('Valor incorrecto')
                    continue
                
                resultado = agenda.buscar_elemento(tarea_buscada, agenda.lista_tareas)
                if resultado:
                    resultado.vencio()
                    prioridad_legible = traducir_prioridad(resultado.prioridad)
                    estado_legible = traducir_estado(resultado.estado)
                    if resultado.tiempo_restante:
                        dias = f"{resultado.tiempo_restante.days} días restantes"
                    else:
                        dias = "Vencida"
                    print(f"Tarea encontrada: {resultado.nombre}\n Descripción:{resultado.descripcion}\n Prioridad:{prioridad_legible}\n Vencimiento: {resultado.fecha_limite.strftime('%d/%m/%Y')}\n Tiempo restante:{dias}\n Estado:{estado_legible}\n -----------------")
                else:
                    print("Tarea no encontrada.")
        
        elif opcion == '5': # Borrar
            while True:
                try:
                    quiere_borrar = input("¿Usted quiere borrar una tarea? [ SI - NO ]")
                    if quiere_borrar.lower() != "si":
                        break
                    tarea_borrar = input('¿Cómo se llama la tarea que quiere borrar?')
                except ValueError:
                    print('Valor incorrecto')
                tarea_borrada = agenda.borrar_elemento(tarea_borrar, agenda.lista_tareas)
                if tarea_borrada:
                    print(f'La tarea borrada es {tarea_borrar}')
                else:
                    print('No se encontro una tarea para borrar')
        
        elif opcion == '6': # Lista Tarea por prioridad
            tareas_x_prioridad = agenda.listar_tareas("prioridad")
            print(f'-------------------------------\nLista de tareas por prioridad\n')
            for t in tareas_x_prioridad:
                t.vencio()
                prioridad_legible = traducir_prioridad(t.prioridad)
                estado_legible = traducir_estado(t.estado)
                if t.tiempo_restante:
                    dias = f"{t.tiempo_restante.days} días restantes"
                else:
                    dias = "Vencida"
                print(f"Nombre:{t.nombre}\n Descripción: {t.descripcion}\n Prioridad: {prioridad_legible}\n Vencimiento: {t.fecha_limite.strftime('%d/%m/%Y')} \n Tiempo restante: {dias}\n Estado: {estado_legible}\n -----------------")
        
        elif opcion == '7': # Lista Tarea por vencimiemto
            tareas_x_vencimiento = agenda.listar_tareas("vencimiento")
            print(f'-------------------------------\nLista de tareas por vencimiento\n')
            for t in tareas_x_vencimiento:
                t.vencio()
                prioridad_legible = traducir_prioridad(t.prioridad)
                estado_legible = traducir_estado(t.estado)
                
                if t.tiempo_restante:
                    dias = f"{t.tiempo_restante.days} días restantes"
                else:
                    dias = "Vencida"
                print(f"Nombre: {t.nombre}\n Descripción: {t.descripcion}\n Prioridad: {prioridad_legible}\n Vencimiento: {t.fecha_limite.strftime('%d/%m/%Y')} \n Tiempo restante: {dias}\n Estado: {estado_legible}\n -----------------")

        
        # Si el usuario elige un accion de eventos :
        elif opcion.lower() == 'a':# Agregar evento                
            while True:
                try:    
                    nombre_evento = input('¿Cómo se llama el Evento?')
                    nombre_existe = False # me evita un bucle infinito
                    for t in agenda.lista_eventos:
                        if t.nombre.lower() == nombre_evento.lower(): # Evita repeticiones
                            nombre_existe = True
                    if nombre_existe: # Idem nombre if nombre_existe == True
                        print('Usted ingreso un nombre que ya exite')
                    else:
                        descripcion = input('Descripción del evento: ')
                        fecha = input('Fecha del evento [dd/mm/aaaa]: ')
                        duracion = input('Duración estimada del evento (en horas): ')
                        ubicacion = input('Ubicación del evento: ')
                        agenda.agregar_evento(nombre_evento, descripcion, fecha, duracion, ubicacion)
                        print("Evento agregado correctamente.")
                        break
                
                except Exception as e:
                    print("Hubo un error al agregar el evento:", e)

        elif opcion.lower() == 'b':  # Modificar evento
            while True:
                try:
                    quiere_modificar = input("¿Usted quiere modificar un evento? [ SI - NO ]")
                    if quiere_modificar.lower() != "si":
                        break
                    evento_modificar = input('¿Cómo se llama el evento que quiere modificar?')
                except ValueError:
                    print('Valor incorrecto')
                    continue
                evento = agenda.buscar_elemento(evento_modificar, agenda.lista_eventos)
                if not evento:
                    print("Evento no encontrado.")
                    break
                while True:
                    otro_atributo = input("¿Usted quiere modificar un atributo? [ SI - NO ]")
                    if otro_atributo.lower() != "si":
                        break
                    atributo = input("¿Qué atributo desea modificar? (nombre, descripcion, ubicacion, fecha, duracion, ): ")
                    nuevo_valor = input("Nuevo valor: ")
                    if atributo == "fecha":  # parche para un error al modificar
                        atributo = "fecha_limite"
                    if atributo == "fecha_limite":
                        try:
                            nuevo_valor = datetime.strptime(nuevo_valor, "%d/%m/%Y")
                        except:
                            print("Fecha inválida. No se modificó.")
                            continue
                    evento.modificar(atributo, nuevo_valor) # Descripcion y  nombre no necesitan ser procesados, asi que se ingresan directamente con el metodo modificar.

        elif opcion.lower() == 'c':  # Marcar evento como asistido
            
            while True:
                try:
                    marcar_completado = input('¿Quiere marcar un evento como asistido? [ SI - NO ]')
                    if marcar_completado.lower() != "si":
                        continue
                    evento_completar = input('¿A qué evento asistió?')
                except ValueError:
                    print('Valor incorrecto')
                    continue
                evento = agenda.buscar_elemento(evento_completar, agenda.lista_eventos) # Si el evento exite la marco como completada.
                if evento:
                    evento.completar()
                    print(f"Usted asistió a el evento {evento.nombre}.")
                else:
                    print(f"Evento {evento_completar} no encontrada.")
                break
        
        elif opcion.lower() == 'd':  # Buscar evento
            while True:
                try:
                    quiere_buscar = input("¿Usted quiere buscar un evento? [ SI - NO ]")
                    if quiere_buscar.lower() != "si":
                        break
                    evento_buscado = input('¿Cómo se llama el evento que quiere buscar?')
                except ValueError:
                    print('Valor incorrecto')
                    continue
                
                evento_x = agenda.buscar_elemento(evento_buscado, agenda.lista_eventos)
                if evento_x:
                    evento_x.vencio()
                    estado_legible = traducir_estado(evento_x.estado)
                    if evento_x.tiempo_restante:
                        dias = f"{evento_x.tiempo_restante.days} días restantes"
                    else:
                        dias = "Vencida"
                    print(f"Evento encontrado:{evento_x.nombre}\n Descripción:{evento_x.descripcion}\n Fecha:{evento_x.fecha_limite.strftime('%d/%m/%Y')}\n Lugar:{evento_x.ubicacion}\n Duración:{evento_x.duracion} horas\n Estado:{estado_legible} \n Tiempo restante:{dias} \n  -----------------")
                else:
                    print("Evento no encontrado.")
        
        elif opcion.lower() == 'e':  # Borrar evento
            while True:
                try:
                    quiere_borrar = input("¿Usted quiere modificar un evento? [ SI - NO ]")
                    if quiere_borrar.lower() != "si":
                        continue
                    evento_borrar = input('¿Cómo se llama el evento que quiere borrar?')
                except ValueError:
                    print('Valor incorrecto')
                    continue
                
                evento_borrado = agenda.borrar_elemento(evento_borrar, agenda.lista_eventos)
                if evento_borrado:
                    print(f'El evento borrado es {evento_borrar}')
                else:
                    print('No se encontro un evento para borrar')

        elif opcion.lower() == 'f':  # Listar eventos por fecha
            print(f'-------------------------------\nLista de eventos por fecha\n')
            eventos = agenda.listar_eventos('fecha')
            for e in eventos:
                e.vencio()
                estado_legible = traducir_estado(e.estado)
                if e.tiempo_restante:
                    dias = f"{e.tiempo_restante.days} días restantes"
                else:
                    dias = "Vencida"
                print(f"Nombre:{e.nombre}\n Descripción:{e.descripcion}\n Fecha:{e.fecha_limite.strftime('%d/%m/%Y')}\n Lugar:{e.ubicacion}\n Duración:{e.duracion} horas\n Estado:{estado_legible} \n Tiempo restante:{dias} \n  -----------------")

        elif opcion.lower() == 'g':  # Listar eventos por ubicación
            eventos = agenda.listar_eventos('ubicacion')
            print(f'-------------------------------\nLista de eventos por ubicacion\n')
            for e in eventos:
                e.vencio()
                estado_legible = traducir_estado(e.estado)
                if e.tiempo_restante:
                    dias = f"{e.tiempo_restante.days} días restantes" # No me gustaba como quedaba el tiempo restante, asi que eso muestra dias nomas
                else:
                    dias = "Vencida"
                print(f"Nombre:{e.nombre}\n Descripción:{e.descripcion}\n Fecha:{e.fecha_limite.strftime('%d/%m/%Y')}\n Lugar:{e.ubicacion}\n Duración:{e.duracion} horas\n Estado:{estado_legible} \n Tiempo restante:{dias} \n  -----------------")

        elif opcion.upper() == 'X':
            print("Gracias por usar la agenda. Hasta pronto.")
            break

        else:
            print("Opción inválida. Intente nuevamente.")
```
